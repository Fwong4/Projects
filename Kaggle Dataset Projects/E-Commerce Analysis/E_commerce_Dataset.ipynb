```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from wordcloud import WordCloud
```

```python
df = pd.read_csv("FashionDataset.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace = True)
```

```python
df.head()
```

```python
df.replace('Nan',np.nan,inplace=True)
df.dropna(axis=0,inplace=True)
```

```python
df['MRP'] = df['MRP'].str.replace('Rs\n','')
df['MRP'] = df['MRP'].astype(int)
df['SellPrice'] = df['SellPrice'].astype(int)
df['Sizes'] = df['Sizes'].str.replace('Size:','')
```

```python
df.describe()
```

```python
text1 = " ".join(i for i in df['Sizes'])
word_cloud1 = WordCloud(collocations = False, background_color = 'white',
                        width = 2048, height = 1080).generate(text1)
word_cloud1.to_file('got.png')
```

```python
plt.imshow(word_cloud1, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
text2 = " ".join(i for i in df['BrandName'])
word_cloud2 = WordCloud(collocations = False, background_color = 'white',
                        width = 2048, height = 1080).generate(text2)
word_cloud2.to_file('got.png')
```

```python
plt.imshow(word_cloud2, interpolation='bilinear')
plt.axis("off")
plt.show()
```
